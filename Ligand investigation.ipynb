```python
import pandas as pd
import numpy as np

hiv = pd.read_csv('hiv-pr.csv')
pepsin = pd.read_csv('pepsinA.csv')

hiv_ligands = list(hiv['Ligand ID'])
pepsin_ligands = list(pepsin['Ligand ID'])

common_ligands = list(set(hiv_ligands).intersection(set(pepsin_ligands)))
common_ligands = [x for x in common_ligands if str(x) != 'nan']
print(common_ligands)

hiv[hiv['Ligand ID'].isin(common_ligands)]
```
